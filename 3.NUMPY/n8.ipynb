```python
import numpy as np
```

```python
a1=np.array([10,20,30])
print(a1)
```

```python
#Copy method (it is independent method)
s_a=np.copy(a1)
print(s_a)
s_a[0]=100
print(s_a)
print(a1)
```

```python
#veiw is dependent and stored in same memory location  
a3=a1.view() 
print('a3:',a3)
print('a3:',a3)

a3[-1]=200
print('a3:',a3)
print('a1:',a1)
```
