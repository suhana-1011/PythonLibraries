```python
import numpy as np
```

```python
a1=np.arange(10,96,5)
print(a1)
print(a1.shape)
```

```python
#reshape is used to convert one type of shape to other 
np.reshape(a1,shape=(2,9))
```

```python
np.reshape(a1,(3,6))
```

```python
a2=np.reshape(a1,(3,2,3))
print(a2)
```

```python
np.reshape(a2,(-1))
```

```python
a3=np.array([[10,20,30,40,50,78],[2,4,5,60,64,28],[11,9,37,92,55,68]])
print(a3)
print(a3.shape)
print(a3.ndim,a3.size,sep='\n')
```

```python
np.reshape(a3,(3,2,1,3))
np.reshape(a3,(3,3,2))

```
