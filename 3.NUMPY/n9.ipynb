```python
import numpy as np
```

```python
a1=np.array([[1,2,3],[10,20,30]])
print(a1)
```

```python
np.reshape(a1,(-1))
```

```python
#ravel function behaves like view method 
# it is going to update the original array also
x=a1.ravel() #->it is coverting in to 1d array
print(x)
x[0]=50
print(a1,x,sep='\n')
```

```python
a4=np.array([[11,22,33],[4,67,80]])
print(a4)
```

```python
#flatten it is coverting in to 1d array
#it is independent it doesnt update the original array
y=a4.flatten() 
print(y)

y[-1]=900
print(a4,y,sep='\n')
```
