```python
lst=[1,2,3,4]
a1=np.array(lst) #for converting list to array
a1
type(a1)
```

```python
list1=[10,2,3.5,7,'np','6j']
a2=np.array(list1) #for converting list to array
type(a2)
print(a2)
# Arrays are homogeneous it cannot store diferent type of data
```

```python
l1=[10,20,30,40,50]
l2=[1,2,3,4,5]
print(l1+l2) #list can concatenate whereas numpy cannot below ex
```

```python
arr1=np.array(l1)
arr2=np.array(l2)
print(arr1+arr2)


#numpy works faster with array not with list 
```
