```python
np.linspace(0,10) #divides the no in to equal parts
np.linspace(0,10,num=5)
np.linspace(0,10,num=5,retstep=True) #shows the difference 
np.linspace(0,10,num=5,retstep=True,endpoint=False)
```

```python
#used for creating num of rows and cols only with 0's matrix
np.zeros(2)
np.zeros((2,3)) 
np.zeros((2,3,4)) 
```

```python
# used for creating num of rows and cols only with 1's by matrix
np.ones(2)
np.ones((2,3))
np.ones((2,3,5))
```

```python
# eye will generate always square matrix i,e identity matrix (diagonal=1)
np.eye(3)
np.eye(4,5)
```

```python
np.full(5,5) #syn:np.full(size,element)
np.full((2,4),100) #2 rows 4 cols with value 100
```

```python
#module random is used to generate random num
import numpy.random as rd
```

```python
# used random integer values within the range of 2
rd.randint(2)
```

```python
rd.randint(10,100) #range between 10-100
rd.randint(10,100,(3)) #it will create in the form matrix with 3 rows
rd.randint(10,100,(3,2)) #it will create in the form matrix with 3rows and 2cols
```

```python
#seed is used to fix the array without changing whenever we run
rd.seed(2)
rd.randint(10,100,(3,2)) # output:array([[50, 25],[82, 32],[53, 92]], dtype=int32)
print(rd.randint(10,100,(3,2))) #print fun is used to avoid dtype
       
```

```python
rd.rand(3) #it gives floating values between 0-1 and only takes size
rd.rand(2,3)
```

```python
r=rd.randn(2)
print(r)

print(np.mean(r))
print(np.std(r))
```

```python
r=rd.randn(2,3)
print(r)

np.mean(r)
np.std(r)
```

```python
#choice method
l=[10,20,30,40,50]
rd.choice(l,3) #it will choose the anyone num from l
```

```python
rd.choice(l,3,replace=False) #use replace to avoid repetition of num
```

```python
l=[10,20,30,40,50]
rd.choice(l,(2,3)) #it will create 2rows and 3 cols
```
