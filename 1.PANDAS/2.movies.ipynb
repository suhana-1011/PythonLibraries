melt():is used to convert col to rows and we use when we want group or visualize
it has 3 parameters
pd.melt(df,id_vars,value_vars,vars_names,value_names)
id_vars=columns doesnt want to change
value_vars=columns to change rows 
var_names=columns name for (old columns)
value_names=columns name for values

```python
pd.melt(df,value_vars=['hotstar','netflix'])
```

```python
pd.melt(df,value_vars=['hotstar','netflix'],var_name='OTT')
```

```python
pd.melt(df,value_vars=['hotstar','netflix'],var_name='OTT',value_name='available')
```

```python
pd.melt(df,id_vars=['Movies','rating'],value_vars=['hotstar','netflix'],var_name='OTT',value_name='available')
```

```python
df1=pd.melt(df,id_vars=['Movies','rating'],value_vars=['hotstar','netflix'],var_name='OTT',value_name='Available')
```

### using aggregate functions(sum, min, max,count,mean)

```python
df1['rating'].count() #using count
```

```python
df1['rating'].agg(['min','max','mean','count']) #using all aggregate fn
```

```python

df1.agg({'rating':'min','Movies':'count'})
```

```python
df1.agg({'rating':['min','mean'],'Movies':'count'})
```
