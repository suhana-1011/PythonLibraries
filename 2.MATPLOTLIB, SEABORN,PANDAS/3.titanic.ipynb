```python
def comp_type(row):
     f=row['fam_count']
     p=row['no_of_psgs']
     if p==1:
          return 'solo'
     elif f==0 and p>1:
          return 'friends'
     else:
          return 'family'
```

```python
df['companion type']=df.apply(comp_type,axis=1)
```

```python
Royal=['the Countees','Jonkheer','Don','Dona','Master','Sir']
Common=['Miss', 'Mr','Mrs','Lady','Ms','Mme','Mlle']
Miltary=['Major','Col','Capt']
other_professional=['Dr','Rev']
```

```python
def social_status(row):
    c=row['Title']
    if c in Royal:
        return 'Royal'
    elif c in Common:
        return 'Common'
    elif c in Miltary:
        return 'Miltary'
    else:
        return 'other professional'
    
```

```python
df['social_status']=df.apply(social_status,axis=1)
```

```python
plt.figure(figsize=(3,4))
cnt=sns.countplot(data=df,x='pclass',color='Black',order=[3,1,2]) # to arrange in order the graph first highest
cnt.bar_label(cnt.containers[0])# to count the pclass containers 
plt.title('No of psngs in each class') #to give title to the y axis
plt.yticks(range(0,900,100)) 
plt.ylabel('No of psngs')
plt.show()#to remove the unwanted text
```

```python
plt.figure(figsize=(8,4))
cnt=sns.countplot(data=df,x='age_category',hue='gender',palette=['cyan','crimson']) # to arrange in order the graph first highest
# cnt.bar_label(cnt.containers[0])
for i in cnt.containers: # use for loop since graph contains two containers at a time
    cnt.bar_label(i)  #then it give count to both containers
plt.title('Distribution of psngs acc to age category') #to give title to the y axis
plt.yticks(range(0,1000,100)) 
plt.ylabel('No of psngs')
plt.show()
```

```python
plt.figure(figsize=(3,4))
c=sns.barplot(data=df,x='gender',y='fare',errorbar=None,estimator='mean') #estimator is to find weather the graph is mean or median by default it is mean
c.bar_label(c.containers[0],fmt='%.1f')
plt.title('Avg Fare gender wise')
plt.yticks(range(0,20,2))
plt.show()
```

***observation***

Average fare is more in female compared to males

```python
plt.figure(figsize=(3,4))
c=sns.barplot(data=df,x='gender',y='fare',errorbar=None,estimator='sum') #estimator is to find weather the graph is mean or median by default it is mean
c.bar_label(c.containers[0],fmt='%.1f')
plt.title('Avg Fare gender wise')
plt.yticks(range(0,14000,2000))
plt.show()

#only mean is changed to sum
```

***Average fare is more in male passengers when compared to male since we have used estimator as sum***

```python
plt.figure(figsize=(5,4))
c=sns.barplot(data=df,x='age_category',y='fare',errorbar=None,
              estimator='mean',hue='gender',palette=['crimson','blue']) 
#hue is used to make subgroups based on gender and pallete is used to change color of two groups 
for i in c.containers:
     c.bar_label(i,fmt='%.1f')
plt.title('Avg Fare by age category')
plt.legend(loc='best',ncol=2,title='gender') #we can give loc as center upperright left best and ncol can be sepreated from 1 below the ther to beside
plt.yticks(range(0,40,5))
plt.show()
#only mean is changed to sum
```

***avergae fare in seniors are more compared to other age category***

```python
g=df.loc[df['companion type']=='family']
gf=g['gender'].value_counts()
gn=g['age_category'].value_counts()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(6,3))
plt.suptitle('companion type wise categorisation')

axes[0].pie(x=gf.values,labels=gf.index,autopct='%.1f%%')
axes[0].set_title('gender wise categorisation')

axes[1].pie(x=gn.values,labels=gn.index,autopct='%.1f%%')
axes[1].set_title('age_category wise categorisation')

plt.tight_layout()#it will adjust overlapping
plt.show()


```
