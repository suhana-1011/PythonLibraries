```python
df=df.reset_index(drop=True)
```

```python
df=df.drop(columns=['store_name','gender_short'])
```

```python
col_lst=['ID','recorddate','DOB','hiredate','termdate','age','experience','city','dname','job',
         'gender','termreason','termtype','Year','Status','Business unit']
df.columns=col_lst
df
```

```python
df['Status'] = df['Status'].str.title()
df['Business unit'] = df['Business unit'].str.title()
df
```

```python
df[[ 'age','experience' ]].agg(['min','max'])
```

```python
df['age category'] = pd.cut(df['age'], bins=[18,25,59,65], labels=['Teen','Adult','Senior'])
df
```

```python
df['service category'] = pd.cut(df['experience'], bins=[-1,1,5,26], labels=['Fresher','Senior','Experts'])
df
```

```python
df['dname'].unique()
```

```python
df['job'].unique()
```

```python
def job_clean(row):
    j=row['job']
    d=row['dname']
    if j!='Auditor' and j!='Dairy Person':
        if d in j:
            return j.replace(d,'').strip().strip(',').strip(', s')
        else:
            return j.removeprefix('Exec Assistant').strip(', ')
    else:
        return j
```

```python
df['job']=df.apply(job_clean,axis=1)
```

```python
df['job'].unique()
```

```python
df['job']=df['job'].replace({'CHief Information Officer':'Chief Information Officer'})
```

```python
df['job'].unique()
```

```python
df.loc[df['Status']=='Active','termdate'].unique()
```

```python
h_df=df['hiredate'].dt.year.value_counts().sort_values().reset_index()
h_df.columns=['year','hiring']
```

```python
t_df=df.loc[df['Status']=='Terminated','termdate'].dt.year.value_counts().sort_values().reset_index()
t_df.columns=['year','termination']
```

```python
df1=pd.merge(h_df,t_df,on='year', how='outer').fillna(0).astype(int)
df1
```

```python
df1['retention1']=df1['hiring']-df1['termination']
df1
```

```python
df1['retention']=df1['retention1'].cumsum()
df1
```

```python
df1=df1.drop(columns='retention1')
df1
```

```python
plt.figure(figsize=(3,4))
sns.countplot(data=df,x='Status')
plt.title('Distribution of emps according to their status')
plt.ylabel('no of emps')
plt.show()
```

```python
c=df['gender'].value_counts()
c.values
c.index
```

```python
plt.figure(figsize=(3,3))
plt.pie(x=c.values, labels=c.index, autopct='%1.1f%%')
plt.title('Distribution of emps according to their gender')
plt.show()
```

*Observation*

The workforce is balanced with slight majority of females than males

```python
plt.figure(figsize=(5,7))
b=sns.barplot(data=df,x='Status',y='experience',errorbar=None, palette=['lightblue', 'lightcoral'],hue='gender')
for i in b.containers:
    b.bar_label(i,fmt='%.1f',padding=-15)
plt.title('Average experience of emps gender wise')
plt.yticks(range(0,15,2))
plt.legend(loc='best',ncol=2,title='Gender')
plt.show()
```

*Observation*

Active emps have highest avg experience than terminated emps

```python
t=df.loc[df['Status']=='Terminated']
tt=t['termtype'].value_counts()
tr=t['termreason'].value_counts()
```
